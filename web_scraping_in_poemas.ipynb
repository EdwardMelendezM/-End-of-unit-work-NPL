## **PREGUNTA 5**: Crear los siguientes módulos independientes para normalizar un texto:

### 1.   Quitar signos de puntuación

```python
# Función que dada una lista de palabras, elimina los signos de puntuación
def remove_punctuation(palabras):
    doc = nlp(" ".join(palabras))
    palabras_sin_puntuacion = [token.text for token in doc if not token.is_punct]    
    return palabras_sin_puntuacion
```

### 2.   Convertir a minúsculas

```python
# Función que dada una lista de palabras, las transforma a minúsculas
def to_lowercase(palabras):
    doc = nlp(" ".join(palabras))
    palabra_lower = [token.text.lower() for token in doc]    
    return palabra_lower
```

### 3.   Quitar stop-words

```python
# Función que dada una lista de palabras, elimina las Stop Words
def remove_stop_words(palabras):
    doc = nlp(" ".join(palabras))
    palabra_is_stop = [token.text for token in doc if not token.is_stop]    
    return palabra_is_stop
```

### 4.   Quitar palabras con extensión mínima (valor por defecto del parámetro: 2)

```python
# Función que dada una lista de palabras y un número mínimo de caracteres que tienen que tener las palabras, 
# elimina todas las palabras que tengan menos caracteres que los indicados
def remove_short_words(palabras,n=2):
    doc = nlp(" ".join(palabras))
    palabras_mayores_n = [token.text for token in doc if len(token.text)>n]    
    return palabras_mayores_n
```

### 5.   Lematizar

```python
# Función que dada una lista de palabras, devuelve esa lista con el lema de cada una de esas palabras
def lemmatizer(palabras):
    doc = nlp(" ".join(palabras))
    palabra_lema = [token.lemma_ for token in doc]
    return palabra_lema
```

### 6. Eliminar duplicados

```python
# Funcion para eliminar palabras duplicadas
def remove_duplicates(palabras):
    p_unicas = set(palabras)   
    return p_unicas
```

### Lista de adjetivos, verbos, sustantivos y adverbios más comunes 

```python
# importamos la libreria FreqDist para  calcular la frecuencia de cada palabra
from nltk.probability import FreqDist

def obtener_palabras_comunes(palabras,nro_comunes):
    # Etiquetar cada palabra con su categoría gramatical
    palabras_etiquetadas = get_PoS(palabras)

    # Filtrar palabras por categoría gramatical
    adjetivos = [palabra[0] for palabra in palabras_etiquetadas if palabra[1]  in ['ADJ']]
    #print(adjetivos)
    verbos = [palabra[0] for palabra in palabras_etiquetadas if palabra[1] in ['VERB']]
    #print(verbos)
    sustantivos = [palabra[0] for palabra in palabras_etiquetadas if palabra[1] in ['NOUN', 'PROPN']]
    #print(sustantivos)
    adverbios = [palabra[0] for palabra in palabras_etiquetadas if palabra[1]in ['ADV']]
    #print(adverbios)

    # Calcular la frecuencia de cada lista de palabras
    fdist_adjetivos = FreqDist(adjetivos)
    fdist_verbos = FreqDist(verbos)
    fdist_sustantivos = FreqDist(sustantivos)
    fdist_adverbios = FreqDist(adverbios)

    # Obtener las palabras más comunes de cada categoría
    adjetivos_comunes = fdist_adjetivos.most_common(nro_comunes)
    verbos_comunes = fdist_verbos.most_common(nro_comunes)
    sustantivos_comunes = fdist_sustantivos.most_common(nro_comunes)
    adverbios_comunes = fdist_adverbios.most_common(nro_comunes)

    return adjetivos_comunes, verbos_comunes, sustantivos_comunes, adverbios_comunes
```

```python
import pandas as pd
def mostrar_tabla_de_poemas(titulos,text_raw):

  def limpiarCadena(cadena):
    caracteres_eliminar = ['\r', '\n', '\xa0']
    for caracter in caracteres_eliminar:
      cadena = cadena.replace(caracter, '')
    return cadena
    
  def limpiarLista(lista):
    for indice in range(len(lista)):
      lista[indice]=limpiarCadena(lista[indice])
    return lista

  def calc_number_parrafos(value):
    parrafos=[]
    parrafosAux = value.split("\n\r\n")
    return len(parrafosAux)
  dicc = obtener_poemas(titulos,text_raw)
  data = {
    'poema': [],
    'num_parrafos': [],
    'oraciones': [],
    'palabras':[]
  }
  for key in dicc:
    num_parrafos = calc_number_parrafos(dicc[key])
    oraciones = segmentar_en_oraciones(dicc[key])
    palabras = get_tokens(dicc[key])
    data['poema'].append(limpiarCadena(dicc[key]))
    data['num_parrafos'].append(num_parrafos)
    data['oraciones'].append(', '.join(limpiarLista(oraciones)))
    data['palabras'].append(', '.join(palabras))
  df = pd.DataFrame(data)
  return df

#-----------------------------------------
#              MOSTRAR TABLA
#-----------------------------------------
mostrar_tabla_de_poemas(titulos,text_raw)
```
